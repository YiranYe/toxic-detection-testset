```python
import demoji
from nltk.corpus import words

word_set = set(words.words())

def simplify_word(word):
    pw = list(word.lower())
    i = 1
    while i < len(pw):
        if pw[i] == pw[i-1]:
            pw.pop(i)
        else:
            i += 1
    return ''.join(pw)


def lowercase_uppercase(pert_word):
    return pert_word.lower() != pert_word


def interesting_lowercase_uppercase(pert_word):
    collect = ''
    for c in pert_word:
        if c.lower() != c:
            collect += c
        
    return pert_word.lower() != pert_word and collect.lower() in word_set and len(collect) != len(pert_word) and len(collect) != 1


def hyphenation(pert_word):
    return '-' in pert_word
    

def emoticons(pert_word):
    return len(demoji.findall(pert_word)) >= 1


def repeat_char(clean_word, pert_word):
    c,p = clean_word.lower(),pert_word.lower()
    simp_c, simp_p = simplify_word(c), simplify_word(p)
    return len(p) > len(c) and simp_c == simp_p and simp_p != p

def abbr(clean_word, pert_word):
    c,p = clean_word.lower(),pert_word.lower()
    i = 0
    for character in c:
        if character == p[i]:
            i += 1
            if i == len(p):
                return len(p) < len(c)
    return False

```

```python
lowercase_uppercase_dict = defaultdict(list)
interesting_lowercase_uppercase_dict = defaultdict(list)
hyphe_dict = defaultdict(list)
emoticons_dict = defaultdict(list)
repeat_char_dict = defaultdict(list)
abbr_dict = defaultdict(list)
other_dict = defaultdict(list)

for k,v in hard_pert.items():
    for val in v:
        if lowercase_uppercase(val):
            lowercase_uppercase_dict['__total__'].append(val)
            lowercase_uppercase_dict[k].append(val)
        if interesting_lowercase_uppercase(val):
            interesting_lowercase_uppercase_dict['__total__'].append(val)
            interesting_lowercase_uppercase_dict[k].append(val)
        if hyphenation(val):
            hyphe_dict['__total__'].append(val)
            hyphe_dict[k].append(val)
        if emoticons(val):
            emoticons_dict['__total__'].append(val)
            emoticons_dict[k].append(val)
        if repeat_char(k,val):
            repeat_char_dict['__total__'].append(val)
            repeat_char_dict[k].append(val)
        if abbr(k,val):
            abbr_dict['__total__'].append(val)
            abbr_dict[k].append(val)
        if not lowercase_uppercase(val) and not hyphenation(val) and not emoticons(val) and not repeat_char(k,val) and not abbr(k,val):
            other_dict['__total__'].append(val)
            other_dict[k].append(val)
```
